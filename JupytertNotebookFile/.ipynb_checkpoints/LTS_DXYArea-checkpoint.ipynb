```python
# github仓库地址
# df_area = pd.read_csv('https://raw.githubusercontent.com/BlankerL/DXY-COVID-19-Data/master/csv/DXYArea.csv')
# gitee上对应的镜像仓库，不可用
# df_area = pd.read_csv('https://gitee.com/Mr_Wr0ng/DXY-COVID-19-Data/raw/master/csv/DXYArea.csv')
df_area = pd.read_csv('G:\Gitclone\DXY-COVID-19-Data\csv\DXYArea.csv')
df_area['date'] = df_area['updateTime'].apply(lambda x: x[:10])
df_area
```

## df_area_simple

### func drop_duplicate_data

```python
def drop_duplicates_data(target_date):
    df_day = df_area[df_area['date'] == target_date]
#     return df_day.drop_duplicates(['countryName', 'provinceName']).sort_values(by=column, ascending=False)
    return df_day.drop_duplicates(['countryName', 'provinceName'])
```

### df_area_simple

```python
df_day_all = [drop_duplicates_data(target_date=d) for d in df_area['date'].unique()]
df_area_simple_t = pd.concat(df_day_all[::-1])
df_area_simple_t['province_nowconfirmedCount'] = df_area_simple_t['province_confirmedCount'] - df_area_simple_t['province_curedCount'] - df_area_simple_t['province_deadCount']
df_area_simple_t['city_nowconfirmedCount'] = df_area_simple_t['city_confirmedCount'] - df_area_simple_t['city_curedCount'] - df_area_simple_t['city_deadCount']
df_area_simple = df_area_simple_t[(df_area_simple_t['countryName'] != '大不列颠及北爱尔兰联合王国') & (df_area_simple_t['countryName'] != '英国（含北爱尔兰）')]
df_area_simple
```

### df_foreign

```python
df_foreign = df_area_simple[df_area_simple['countryName'] != '中国']
df_foreign
```

#### dff_forergn

```python
dff_foreign = pd.pivot(data=df_foreign, index='date', columns='countryName', values=['province_confirmedCount',
                                                                                     'province_curedCount',
                                                                                     'province_deadCount', 
                                                                                    'province_nowconfirmedCount'])
dff_foreign.iloc[0, :].fillna(0, inplace=True)
dff_foreign.fillna(method='ffill', inplace=True)
dff_foreign.columns.names = ['countType', 'countryName']
dff_foreign
```

### df_cn

```python
df_cn = df_area_simple[(df_area_simple['countryName'] == '中国') & (df_area_simple['provinceName'] != '中国')]
df_cn['provinceShortName'] = df_cn['provinceName'].apply(lambda x: re.sub('[市,省,自治区,壮族,回族,维吾尔]', '', x))
df_cn
```

### df_province

```python
df_province = pd.pivot(data=df_cn, index='date', columns='provinceShortName', values=['province_confirmedCount', 
                                                                                 'province_curedCount',
                                                                                 'province_deadCount', 
                                                                                'province_nowconfirmedCount'])
df_province.iloc[0, :].fillna(0, inplace=True)
df_province.fillna(method='ffill', inplace=True)
df_province.columns.names = ['countType', 'provinceShortName']
df_province
```

```python
dff_province = df_province
```

### dff_cn

```python
def df_cn_count(column):
    df = df_province[column]
    return df.sum(axis=1)
df_cn_columns = [['province_confirmedCount', 'province_curedCount', 'province_deadCount', 'province_nowconfirmedCount'], 
                 ['中国', '中国', '中国', '中国']]
df_cn_sum = [df_cn_count(column=c) for c in df_cn_columns[0]]
dff_cn = pd.concat(df_cn_sum, axis=1)
dff_cn.columns = df_cn_columns
dff_cn.columns.names = ['countType', 'countryName']
dff_cn
```

## df_world

```python
df_world = pd.concat([dff_cn, dff_foreign], axis=1,sort=False)
# df_world.iloc[0,:].fillna(0, inplace=True)
# df_world.fillna(0, inplace=True)
df_world.fillna(method='ffill', inplace=True)
df_world.fillna(0, inplace=True)
df_world
```

### func melt_data

```python
dff_world['confirmedCount'].melt(value_name='confirmedCount')
def melt_data(column):
    df = dff_world[column]
    df['date'] = df.index
    dff = df.melt( id_vars='date',value_name=column)
    return dff
# melt_data(column='nowconfirmedCount')
# melt_data_ls = [melt_data(c) for c in ['confirmedCount', 'nowconfirmedCount', 'curedCount', 'deadCount']]
# dff_world_melt = pd.concat(melt_data_ls, axis=1)
# dff_world_melt
```

### merge_all

```python
melt_data_confirmed = melt_data(column='confirmedCount')
melt_data_nowconfirmed = melt_data(column='nowconfirmedCount')
melt_data_cured = melt_data(column='curedCount')
melt_data_dead = melt_data(column='deadCount')

merge_con_nowcon = pd.merge(left=melt_data_confirmed, right=melt_data_nowconfirmed, on=['date', 'countryName'])
merge_cured_dead = pd.merge(left=melt_data_cured, right=melt_data_dead, on=['date', 'countryName'])
merge_all = pd.merge(left=merge_con_nowcon, right=merge_cured_dead, on=['date', 'countryName'])
merge_all.sort_values(by='date', inplace=True)
merge_all
```

```python
df_countryNameDiff = pd.read_excel('ccname.xlsx', sheet_name='Sheet3')
countryNameDict = dict(zip(df_area['countryName'], df_area['countryEnglishName']))
countryNameDictDiff = dict(zip(df_countryNameDiff['cn'], df_countryNameDiff['en']))
def new_countryEnName(name):
    if name in countryNameDictDiff:
        return countryNameDictDiff[name]
    return countryNameDict[name]
```

## dff_area

```python
# merge_all['countryEnglishName'] = merge_all['countryName'].apply(lambda x: countryNameDict.get(x))
# merge_all
dff_area = merge_all.reset_index().drop(columns=['index'])
dff_area['countryShortEnglishName'] = dff_area['countryName'].apply(new_countryEnName)
dff_area['countryEnglishName'] = dff_area['countryName'].apply(lambda x: countryNameDict[x])
dff_area.fillna('not value', inplace=True)
dff_area['deadRate'] = dff_area['deadCount'] / dff_area['confirmedCount'] * 100
dff_area['curedRate'] = dff_area['curedCount'] / dff_area['confirmedCount'] *100
dff_area
```

### world_map_tap

```python
from pyecharts.charts import Map, Timeline, Line, Tab
from pyecharts import options as opts
def get_world_map_all(label, column, pieces_type):
    tl = Timeline(init_opts=opts.InitOpts(width='1500px', height='700px'))
    
    tl.add_schema(play_interval=1000)
    
    for target_date in dff_area['date'].unique():
        confirm_map =  Map()
        df_world = dff_area[dff_area['date']==target_date].sort_values(by=column, ascending=False)
        df_world_10 = df_world[df_world[column] > 0].head(30)
#         subtitle_info = '\n'.join([f"{j+1}. {i[0]}: {i[1]:,.0f}" for j,i in enumerate(zip(df_world_10['countryName'], df_world_10[column]))])

        subtitle_info = '\n'.join([f"{i[0]}: {i[1]:,.0f}" for i in zip(df_world_10['countryName'], df_world_10[column])])
        if column == 'deadCount':
            subtitle_info = '\n'.join([f"{i[0]}: {i[1]:,.0f} ({i[2]:.2f}%)" for i in zip(df_world_10['countryName'], df_world_10[column], df_world_10['deadRate'])])
        if column == 'curedCount':
            subtitle_info = '\n'.join([f"{i[0]}: {i[1]:,.0f} ({i[2]:.2f}%)" for i in zip(df_world_10['countryName'], df_world_10[column], df_world_10['curedRate'])])                
        titles = f"{target_date[-5:]} {label}病例: {df_world[column].sum():,.0f}人\n\n{subtitle_info}"
        confirm_map.add(label, 
                    [list(z) for z in zip(df_world['countryShortEnglishName'], df_world[column].tolist() )], 
                    "world",label_opts=opts.LabelOpts(is_show=False),is_map_symbol_show=False,
                  )

        confirm_map.set_global_opts(
            title_opts=opts.TitleOpts(title=titles,
                                      pos_left='right',
                                      ), 
            legend_opts = opts.LegendOpts(pos_bottom='15%', pos_left='40%', is_show=False),
        
            visualmap_opts=opts.VisualMapOpts(is_piecewise=True, 
                                             pieces=pieces_type, 
                                             pos_bottom='10%',
                                             pos_left='15%')
        )
        tl.add(confirm_map, target_date)  
    return tl
pieces_confirm = [{'min': 100000, 'color': '#4F070D'},
                  {'min':50000, "max": 99999, 'color': '#811C24'},
                  {'min':10000, "max": 49999, 'color': '#CB2A2F'},
                  {'min':1000, "max": 9999, 'color': '#E55A4E' },
                  {'min':100, "max": 999, 'color': '#F59E83'},
                  {'min':1, "max": 99, 'color': '#FDEBCF'},
                  {'max':0, 'color': '#FFFFFF', 'label': '0'}]
pieces_cured = [{'min': 10000, 'color': '#25BB00'},
                {'min':1000, "max": 9999, 'color': '#33FF00'},
                {'min':500, "max": 999, 'color': '#5CFF33'},
                {'min':100, "max": 499, 'color': '#92FF77' },
                {'min':10, "max": 99, 'color': '#D6FFCC'},
                {'min':1, "max": 9, 'color': '#E4FFDD'},
                {'max':0, 'color': '#FFFFFF', 'label': '0'}]
pieces_dead = [{'min': 10000, 'color': '#4F070D'},
                {'min':1000, "max": 9999, 'color': '#811C24'},
                {'min':500, "max": 999, 'color': '#CB2A2F'},
                {'min':100, "max": 499, 'color': '#E55A4E' },
                {'min':10, "max": 99, 'color': '#F59E83'},
                {'min':1, "max": 9, 'color': '#FDEBCF'},
                {'max':0, 'color': '#FFFFFF', 'label': '0'}]

world_tab = Tab()
world_tab.add(get_world_map_all(label='现有确诊', column='nowconfirmedCount', pieces_type=pieces_confirm), '世界现有确诊病例疫情图')
world_tab.add(get_world_map_all(label='累计确诊', column='confirmedCount', pieces_type=pieces_confirm), '世界累计确诊病例疫情图')
world_tab.add(get_world_map_all(label='累计治愈', column='curedCount', pieces_type=pieces_cured), '世界累计治愈病例疫情图')
world_tab.add(get_world_map_all(label='累计死亡', column='deadCount', pieces_type=pieces_dead), '世界累计死亡病例疫情图')
world_tab.render('world_tab.html')
```
